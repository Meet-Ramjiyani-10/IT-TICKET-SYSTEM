```python
incident_df["made_sla"].value_counts(normalize=True)
```

very slightly imbalanced 63 sla are met and 37 are breached making sla meaning full (prediction)

```python
incident_df.groupby("priority")["made_sla"].value_counts(normalize=True)
incident_df.groupby("impact")["made_sla"].value_counts(normalize=True)
incident_df.groupby("urgency")["made_sla"].value_counts(normalize=True)
```

SLA breach rates vary across priority, impact, and urgency levels, indicating that operational features available at ticket creation time carry predictive signal without deterministically defining the SLA outcome.

```python
top_categories = incident_df["category"].value_counts().head(10).index

incident_df[
    incident_df["category"].isin(top_categories)
].groupby("category")["made_sla"].value_counts(normalize=True)
```

Different incident categories exhibit varying SLA breach rates, suggesting category-level complexity differences.

## EDA Conclusion
SLA breach prediction is a legitimate ML task in this dataset.
Structured operational features (priority, impact, urgency, category, time) show meaningful variation with SLA outcomes.
No post-resolution features are required to justify initial modeling.
