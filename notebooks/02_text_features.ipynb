```python
import pandas as pd
```

```python
incident_df=pd.read_csv("D:\\PROJECT\\IT SERVICE\\data\\processed\\ucl_incident_master.csv")
```

```python
incident_df["symptom_len"] = incident_df["u_symptom"].astype(str).str.len()

incident_df["symptom_len"].describe()
```

Most symptom descriptions are short, indicating limited but potentially useful textual signal.

```python
incident_df["u_symptom"].value_counts().head(10)
```

A small number of recurring symptom templates exist, suggesting partial standardization.

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words="english", max_features=20)
cv.fit(incident_df["u_symptom"].astype(str))

cv.get_feature_names_out()
```
